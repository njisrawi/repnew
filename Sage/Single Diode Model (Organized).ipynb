# Single Diode [Jain & Kapoor]

```python
from scipy.constants import e, k
from pylab import plot, show, xlabel, ylabel, title, grid

# Initialization
I_min = -0.1028 # Initial current value for voltage evaluation
I_max = 0. # Final current value for voltage evaluation
Steps = 1000 # Total number of evaluation points for plotting

# Constants
q = e # Elementary Charge
kB = k # Boltzmann Constant

# Blue solar cell parameters
Iph = 0.1023 # Photogenerated Current (Amps)
I0 = 0.1036E-6 # Diode Reverse Saturation Current (Amps)
n = 1.5019 # Ideality Factor
Rs = 0.06826 # Series Resistance (Ohms)
Rp = 1000. # Shunt Resistance (Ohms)
T = 300. # Temperature (K)

# Equation
I = var('I')
V =   I*Rs \
    + (I+Iph+I0)*Rp \
    - (n*kB*T)/q * lambert_w(q/(n*kB*T) * I0*Rp * (q/(n*kB*T) * Rp*(I+Iph*I0)).exp() )

# Ploting
x_axis = []
y_axis = []
for j in range(Steps+1) :
    I_step = I_min + j*(I_max-I_min)/Steps
    y_axis.append(I_step)
    x_axis.append(V(I_step))
plot(x_axis,y_axis)
title('I-V Characteristics')
xlabel('V (Volts)')
ylabel('I (Amps)')
grid(True)
show()
```

# Single Diode [Valluri & Roberts]

```python
from scipy.constants import e, k
from pylab import plot, show, xlabel, ylabel, title, grid

# Initialization
I_min = -0.1028 # Initial current value for voltage evaluation
I_max = 0. # Final current value for voltage evaluation
Steps = 1000 # Total number of evaluation points for plotting

# Constants
q = e # Elementary Charge
kB = k # Boltzmann Constant

# Blue solar cell parameters
Iph = 0.1023 # Photogenerated Current (Amps)
I0 = 0.1036E-6 # Diode Reverse Saturation Current (Amps)
Rs = 0.06826 # Series Resistance (Ohms)
Rp = 1000. # Shunt Resistance (Ohms)
T = 300. # Temperature (K)
n = 1.5019 # Ideality Factor

# Equation
I = var('I')
V=  + I*Rs \
    + ((n*kB*T)/q) * log(lambert_w((((q/(n*kB*T)) * I0 * Rp).log() + (q/(n*kB*T)) * Rp * (I+Iph+I0)).exp())) \
    - ((n*kB*T)/q) * log( (q/(n*kB*T)) * I0 * Rp )

# Ploting
x_axis = []
y_axis = []
for j in range(Steps+1) :
    I_step = I_min + j*(I_max-I_min)/Steps
    y_axis.append(I_step)
    x_axis.append(V(I_step))
plot(x_axis,y_axis)
title('I-V Characteristics')
xlabel('V (Volts)')
ylabel('I (Amps)')
grid(True)
show()
```
